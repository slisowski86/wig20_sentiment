```python
#Leave news only with max 125 words in news
news_df['tokens']=news_df['news_text'].apply(lambda x : str(x).split(' '))
```

```python
news_df=news_df[news_df['tokens'].apply(lambda x : len(x))<125]
```

```python
#translate news to English
from tqdm import tqdm
import translators as ts
translated_news=[]
for i in tqdm(range(len(news_df))):
    
    translation=ts.google(str(news_df['news_text'].iloc[i]))
    translated_tup=(news_df['company'].iloc[i],news_df['news_date'].iloc[i], translation)
    translated_news.append(translated_tup)
    
    if(i%300==0)&(i!=0):
        translated_df=pd.DataFrame(translated_news, columns=['company','news_date','news_text'])
        translated_df.to_csv('/home/slisowski/Portfolio/wig20_sentiment/translated_files_prod/embed/news_tr_'+str(news_df.index[i]))
        translated_news=[]
    elif(i==len(news_df)-1):
        translated_df=pd.DataFrame(translated_tup, columns=['company','news_date','news_text'])
        translated_df.to_csv('/home/slisowski/Portfolio/wig20_sentiment/translated_files_prod/embed/news_tr_'+str(news_df.index[i]))
        break
```
